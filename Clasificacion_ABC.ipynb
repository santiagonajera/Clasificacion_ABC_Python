```python
import pandas as pd

# Cargar el archivo Excel desde el repositorio
file_path = 'https://github.com/santiagonajera/Clasificacion_ABC_Python/raw/main/ventas-2024-forecast.xlsx'

# Cargar las hojas necesarias del archivo Excel
df_forecast = pd.read_excel(file_path, sheet_name='Forecast')
df_precios_costos = pd.read_excel(file_path, sheet_name='Precios-Costos')

# Fusionar los dataframes Forecast y Precios-Costos usando la columna 'SKU'
merged_df = pd.merge(df_forecast, df_precios_costos[['SKU', 'Precio']], on='SKU')
merged_df 
```

```python
# Seleccionar las columnas de pronósticos para multiplicarlas por el precio
forecast_columns = df_forecast.columns[1:]  # Selecciona todas las columnas de forecast (de ene-25 a dic-25)
forecast_columns
```

```python

# Multiplicar los valores de pronósticos por el precio para cada mes
for column in forecast_columns:
    merged_df[column] = merged_df[column] * merged_df['Precio']

merged_df
```

```python
# Calcular el total de ventas proyectadas para cada SKU sumando las columnas de forecast
merged_df['Total Ventas Proyectadas'] = merged_df[forecast_columns].sum(axis=1)

merged_df
```

```python
# Crear un dataframe final con los SKU y sus ventas proyectadas totales
result_df = merged_df[['SKU', 'Total Ventas Proyectadas']]
result_df
```

```python
# Ordenar el dataframe por la columna 'Total Ventas Proyectadas' de mayor a menor
result_df = result_df.sort_values(by='Total Ventas Proyectadas', ascending=False)
result_df
```

```python
# Calcular el total de todas las ventas proyectadas
total_sales = result_df['Total Ventas Proyectadas'].sum()

# Calcular el porcentaje acumulado de las ventas
result_df['% Acumulado'] = result_df['Total Ventas Proyectadas'].cumsum() / total_sales * 100

# Clasificación ABC basada en el porcentaje acumulado
def clasificar_abc(porcentaje_acumulado):
    if porcentaje_acumulado <= 60:
        return 'A'
    elif porcentaje_acumulado <= 80:
        return 'B'
    else:
        return 'C'

# Aplicar la clasificación ABC
result_df['Clasificación ABC'] = result_df['% Acumulado'].apply(clasificar_abc)

# Mostrar el resultado final con la clasificación ABC
result_df
```
